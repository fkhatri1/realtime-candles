```python
from FTXWebsocket import FTXWebsocket
import json

def on_message(ws, msg):
    msg_data = json.loads(msg)
    if msg_data['type'] == 'update':
        for raw_trade in msg_data['data']:
            trade = {'time': raw_trade['time']}
            print(trade['price'], trade)
            
def on_close(ws):
    print("CLOSED")

ws = FTXWebsocket(channel="trades", market="BTC-PERP", on_message = on_message, on_close = on_close)
ws.run()
```
